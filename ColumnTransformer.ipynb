```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 0:16], df.iloc[:, -1], test_size=0.2)
```

# Using Column Transformer for encode all of these categorical Columns in A single time

```python
#Gender OneHot
#Age Group OneHot
#Race OneHot
#training ordinal
#reactions ordinal
#Qualified Label
```

```python

transformer = ColumnTransformer(transformers=[
    ('trf1', OneHotEncoder(sparse_output=False, drop='first'), ["Gender", "Age Group", "Race"]),
    ('trf2', OrdinalEncoder(categories=[["Slow", "Average", "Fast"]]), [ "Reactions"])
], remainder = "passthrough")
# remainder= "drop"
```

```python
transformer.fit(X_train)
```

```python
original_index = X_train.index
```

```python
col_name = transformer.get_feature_names_out()
```

```python
# Clean them (remove prefixes like trf1__ or trf2__)
clean_names = [name.split("__")[-1] for name in col_name]
```

```python
X_train_new = transformer.transform(X_train)
```

```python
X_train_new = pd.DataFrame(X_train_new,
columns = clean_names, index = original_index)
```

```python
X_train_new.head()
```
